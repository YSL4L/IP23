Based on the homework description, I am going to use function pos_tag(), word_tokenize() from nltk and pyplot from matplotlib in the following code. 

As for word_tokenize() and pos_tag(), they are respectively responsible for dividing the sentences into words and specifying the Part-of-Speech tags for each word.
```python
from nltk import pos_tag, word_tokenize
import matplotlib.pyplot as plt
```

## nltk - Tagging the words

First I read out the file which needed to be tagged, here I removed the newlines.

```python
with open('data/ice_man.txt', encoding='utf-8') as f:
    full_text = f.read()
print(full_text.replace('\n',' '))
```

The word_tokenize function could split the sentences(strings) into words and store its components (including punctuation)  as substrings in a list, which I named as "tokenized_text".

Syntax is：word_tokenize(s), in which s is a string.

```python
tokenized_text = word_tokenize(full_text)
print(tokenized_text)
```

The pos_tag function would annotate each split component from the list above using POS tags like NN, NNP, VBD, etc. and form a tuple containing two variables, which form a new list that I named "words_and_tags".

Syntax is：pos_tag(w), in which w is a list of subsrtings for tagging.

More information and the whole list of POS tags could be found on https://www.sketchengine.eu/tagsets/.  

```python
words_and_tags = pos_tag(tokenized_text)
print(words_and_tags)
```

Since I would like to get the result that focus on different tags of words, I have created the empty list "all_tags" to collect every single tag and the empty list "word_tags" to collect distinct tag names without duplication, in order to identify and remove punctuation during the counting process.

```python
all_tags = []
for word,tag in words_and_tags:
    all_tags.append(tag)
word_tags = list(set(all_tags))
print (word_tags)
len(word_tags)
```

```python
count_dict = dict()
for word,tag in words_and_tags:
    if tag in count_dict:
        count_dict[tag] += 1
    elif tag == ',' or tag =='.' or tag ==':':
        continue
    else:
        count_dict[tag] = 1
        
print(count_dict)
len(count_dict)
```

In this part, I eventually got a dictionary with keys and values of POS tags and their counts respectively.

## matplotlib - Creating a bar chart

The functions I used here handle the data and display of the chart. For "plt.bar()" and "plt.text()", they respectively use data from "count_dict" to create the x-axis and y-axis, and to add text to the bars in order to make the information more easily to perceive. The rest of funtions could make the chart look neater, descriptions were added as well.

Syntax of some key functions are:

bar(): matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, \*, align='center', data=None, \**kwargs)

text(): matplotlib.pyplot.text(x, y, s, fontdict=None, **kwargs)

```python
plt.figure(figsize=(18, 9))
plt.bar(count_dict.keys(), count_dict.values())
for k,v in count_dict.items():
    plt.text(k,v,s=v, ha = 'center')

plt.title('word usage of text \"ice man\"')
plt.xlabel("tag name")
plt.ylabel('times')
plt.show()
```

The results are shown in the bar chart above.
