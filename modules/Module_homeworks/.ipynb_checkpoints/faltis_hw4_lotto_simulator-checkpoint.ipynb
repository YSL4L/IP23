# <font color=#104E8B>Lotto simulator for homework 4</font>

Write a simple lotto simulator. This should:
- ask the user for a lotto tip (6 different numbers),
- simulate 1 million lotto draws by random generator (6 different numbers between 1 and 45)
- count how often the user's tip would have won or not won (not won, 3s, 4s, 5s, 6s)
- Output these numbers


#### <font color=#1C86EE>Task 1</font>

I decided to code two functions for this homework. One for the user input and the second for the random generated one million numbers and the winnings.
For generating numbers randomly the import of module <a href="https://www.w3schools.com/python/module_random.asp">random</a> is necessary wich  is a build-in module and contains in the python standard library.

```python
import random #import the module random 

def userInput():
    while True: #while loop for input until the condition is not true anymore, so it runs until the user typed six differnt numbers separetet with whitespaces
        numbers = []  
        inputNumbers = input("Please insert 6 numbers seperated by whitespaces: ") #variable containing the input numbers
        numsSplitted = inputNumbers.split() #input numbers get splitted
        for num in numsSplitted: #for loop to append the numbers 
            numbers.append(int(num))
        numbersInSet = set(numbers) #casting the numbers into a set 
        if len(numbersInSet) == 6: #use the len function for counting the numbersInSet and abort condition for while loop 
            return numbers
        else:
            print("Not enough numbers given or duplicated numbers were given!")
```

#### <font color=#1C86EE>Task 2, 3 and 4</font>

This function contains the randomly genarted one million numbers and outputs the number of winnings. The function <a href="https://www.w3schools.com/python/ref_random_sample.asp">sample</a> of the imported module random is used with the parameters range(1,46), 6 for generating 6 random numbers in the range between 1 and 45. In the code it is defined range(1,46) because the last number 46 is not used as random number.


```python
def lotto(userNumbers): #this function runs with the value userNumbers generated by the first function
    wins = {'6s':0, '5s':0, '4s':0, '3s':0, 'not_won':0} #here I used a dictionary because I want use the keys and values to store the winnings
    for draw in range(1000000): #for loop to generate the random 1 million lotto drawings
        drawnNumbers = [] 
        drawnNumbers = random.sample(range(1,46), 6) #here the module random with the function sample is in use to generate six differnt numbers between 1 and 46
        countRightNumbers = 0 
        for num1 in userNumbers: #nested for loops to iterate and compare the userNumbers (input from the user) and the drwanNumbers (randomly generated numbers)
            for num2 in drawnNumbers: 
                if num1 == num2: 
                    countRightNumbers +=1 #variable for right number is increased with 1
            
        if countRightNumbers == 6 or countRightNumbers == 5 or countRightNumbers == 4 or countRightNumbers == 3: 
            wins[str(countRightNumbers)+'s'] +=1 #if condition to update the value(s) to the right key in the dictonary - countingRightNumbers must be minimum 3 to get stored
        else:
            wins['not_won'] +=1 #else condition to update the not won key-value-pair in the dictonary if countingRightNumbers < 3
             
        
    print(f" Your winnings -> 6s: {wins['6s']}, 5s: {wins['5s']}, 4s: {wins['4s']}, 3s: {wins['3s']}, not won: {wins['not_won']}")   
```

call the functions



```python
userNumbers = userInput()
lotto(userNumbers)
        
```
