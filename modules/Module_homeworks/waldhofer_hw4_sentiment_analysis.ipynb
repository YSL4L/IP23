At first we need to import our two modules here we need specific functions of our libraries and use 'from' to get those

```python
from textblob import TextBlob
from matplotlib import pyplot as plt
```

Here we're setting our text variable and read the provided text file

```python
text = ""

with open("ice_man.txt", 'r', encoding='UTF-8') as fh:
    text = fh.read()
```

Now we define the TextBlob object and check if the sentences are tokenized correctly

```python
tb = TextBlob(text)
tb.sentences
```

Here we create a list to where the polarity values are added

```python
sentence_group = []

for sentence in tb.sentences:
    sentence_group.append(sentence.sentiment.polarity)

print(sentence_group)
```

Finally we draw the Barplot using the matplotlib module. Three variables are declared for correctly displaying the values in the plot. We loop through out polarity value list to add to our values if a certain threshold is hit

```python
negative = 0
positive = 0
neutral = 0

for polarity in sentence_group:
    if polarity == 0:
        neutral += 1
    if polarity > 0:
        positive +=1
    if polarity < 0:
        negative +=1

data = (negative,positive,neutral)

columns = ('Negative','Positive', 'Neutral')

plt.bar(columns, data)
```
